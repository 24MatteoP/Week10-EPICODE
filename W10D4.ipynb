```python
#ESERCIZIO
#DATASET INSURANCE CONTIENE DATI RISPETTO A CARATTERISTICHE E ABITUDINI DELLE PERSONE, E DELLA ZONA IN CUI VIVONO, RISPETTO A COSTI INDIVIDUALI
#PER LE CURE MEDICHE COME PREMIO PER LE ASSICURAZIONI SULLA SALUTE

import pandas as pd
url = r"C:\Users\Matteo Perrotta\Downloads\archive\beginner_datasets\insurance.csv"
ins = pd.read_csv(url)
ins
```

```python
#1. VISUALIZZIAMO LE DIMENSIONI
print("righe:",ins.shape[0])
print("colonne:",ins.shape[1])
dati_tot = ins.shape[0] * ins.shape[1]
print("Dati totali sono:",dati_tot)
```

```python
#1. OSSERVIAMO I NOMI DI COLONNA
ins.head(0)
```

```python
#2. QUALI SONO LE MEDIE DI CHARGES RISPETTO A REGION?
ins.groupby('region')['charges'].mean()
```

```python
#3. QUALI SONO LE MEDIE DI CHARGES RISPETTO A SMOKER?

ins['smoker_num'] = ins['smoker'].map({'yes': 1, 'no': 0})
ins.groupby('region')['smoker_num'].mean()
```

```python
#3. QUALI SONO LE MEDIE DI CHARGES RISPETTO A SEX?
ins['sex_num'] = ins['sex'].map({'male': 1, 'female': 0})
ins.groupby('region')['sex_num'].mean()
```

```python
#4. QUALI SONO I DESCRITTORI STATISTICI DI BMI?

descrittori = ins['bmi'].describe()
print(descrittori)
```

```python
#4. QUALI SONO MINIMO, MEDIA E MASSIMO DI CHARGES RISPETTO AI DIVERSI QUARTILI DEI VALORI DI BMI?

ins['bmi'].quantile([0.25, 0.5, 0.75])
q1 = ins['bmi'].quantile(0.25)
q2 = ins['bmi'].quantile(0.5)
q3 = ins['bmi'].quantile(0.75)

def bmi_quartile(bmi):
    if bmi <= q1:
        return 'Q1'
    elif bmi <= q2:
        return 'Q2'
    elif bmi <= q3:
        return 'Q3'
    else:
        return 'Q4'

ins['bmi_quartile'] = ins['bmi'].apply(bmi_quartile)

risultato = ins.groupby('bmi_quartile')['charges'].agg(['min', 'mean', 'max'])
print(risultato)
```
